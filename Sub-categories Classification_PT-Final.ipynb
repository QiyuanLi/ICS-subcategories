```python
def plot_categories(series,title):
    y1=series.value_counts().values/sum(series.value_counts().values)
    x1=range(len(y1))
    fig, ax = plt.subplots(figsize=(len(y1),5))
    rects1 = ax.bar(x1, y1, color='r')
    ax.set_xticks(np.array(x1))
    ax.set_xticklabels(series.value_counts().index,rotation=45)
    ax.set_title(str(title))
    plt.show()
    return
```

```python
def pre_X(X):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(X)
    #tokenizer.num_words=1000
    max_length = max([len(s.split()) for s in X])
    X_tokens = tokenizer.texts_to_sequences(X)
    X_pad = pad_sequences(X_tokens, maxlen = max_length, padding = 'post')
    #X_pad = pad_sequences(X_tokens, maxlen = m_len, padding = 'post')
    vocab_size = len(tokenizer.word_index)+1#take tokenizer out from function
#    return X_pad, vocab_size, max_length
    return X_pad, vocab_size, max_length
```

```python
def create_model(num_hidden_layers = 1, 
                 hidden_size=32, 
                 activation='relu',
                 kernel_init=keras.initializers.RandomNormal(mean=0.0, stddev=0.05, seed=None),
                 dr = 0,
                 l1_num = 0
                ):
    
    input_dim = len(X_train[0])
    model0 = Sequential()
    model0.add(Dense(hidden_size, activation=activation, input_dim=input_dim,
                     kernel_initializer=kernel_init
                    ))
    for i in range(num_hidden_layers):
        model0.add(Dense(hidden_size, 
                         kernel_initializer=kernel_init,
                         activation=activation,
                         activity_regularizer=l1(l1_num),
#                          kernel_constraint=max_norm(4), 
#                          bias_constraint=max_norm(4)
                        ))
        model0.add(Dropout(dr))

    model0.add(Dense(len(y_train[0]),
                     kernel_initializer=kernel_init, 
                     activation='softmax'))
#     Adagrad = keras.optimizers.Adadelta(lr=0.1, rho=0.95, epsilon=None, decay=0.0)
    model0.compile(loss='categorical_crossentropy', optimizer = 'Adam', metrics=['accuracy'])
    return model0
```

```python
def create_model(num_layers=1):
    #parameters#################################################################################################################
    num_hidden_layers = 1
    hidden_size=512
    activation='tanh'
    keras_initializer = keras.initializers.he_normal(seed=None)
    activity_regularizer=l1(1e-11)
    kernel_constraint=max_norm(4)
    bias_constraint=max_norm(4)
    drop_rate = 0.6
    ############################################################################################################################

    #optimizer##################################################################################################################
    lr=0.1
    rho=0.95
    epsilon=None
    decay = 0.0
    Adagrad = keras.optimizers.Adadelta(lr=lr, rho=rho, epsilon=epsilon, decay=decay)
    optimizer = 'Adadelta'
    ############################################################################################################################

    #Model######################################################################################################################
    input_dim = len(X_train[0])

    model0 = Sequential()
    model0.add(Dense(hidden_size, activation=activation, input_dim=input_dim,
                     kernel_initializer=keras_initializer
                    ))
    for i in range(num_hidden_layers):
        model0.add(Dense(hidden_size, 
                         kernel_initializer=keras_initializer,
                         activation=activation,
                         activity_regularizer=activity_regularizer,
                         kernel_constraint=kernel_constraint, 
                         bias_constraint=bias_constraint))
        model0.add(Dropout(drop_rate))

    model0.add(Dense(len(y_train[0]),
                     kernel_initializer=keras_initializer, 
                     activation='softmax'))

    model0.compile(loss='categorical_crossentropy', optimizer = optimizer, metrics=['accuracy'])
    return model0
```

```python
def create_model(num_hidden_layers = 1, 
                 hidden_size=256, 
                 activation='tanh',
                 kernel_init=keras.initializers.he_normal(seed=None),
                 dr = 0.6,
                 l1_num = 1e-11,
                 weight = 4
                ):
    
    input_dim = len(X_train[0])
    model0 = Sequential()
    model0.add(Dense(hidden_size, activation=activation, input_dim=input_dim,
                     kernel_initializer=kernel_init
                    ))
    for i in range(num_hidden_layers):
        model0.add(Dense(hidden_size, 
                         kernel_initializer=kernel_init,
                         activation=activation,
                         activity_regularizer=l1(l1_num),
                         kernel_constraint=max_norm(weight), 
                         bias_constraint=max_norm(weight)
                        ))
        model0.add(Dropout(dr))

    model0.add(Dense(len(y_train[0]),
                     kernel_initializer=kernel_init, 
                     activation='softmax'))
    Adagrad = keras.optimizers.Adadelta(lr=0.1, rho=0.95, epsilon=None, decay=0.0001)
    model0.compile(loss='categorical_crossentropy', optimizer = Adagrad, metrics=['accuracy'])
    return model0
```
